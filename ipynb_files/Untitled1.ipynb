```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation

A = 1
m = 1
g = 9.8
length = 1

dt = 0.01

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.set_aspect('equal', adjustable='box')
ax2.set_aspect('equal', adjustable='box')

ax1.set(xlim=(-1.5 * A,1.5 * A),ylim=(-1.5* A,1.5 * A))

pen_x, = ax1.plot([], [], 'ko')
rod, = ax1.plot([], [], 'k-')
x = []
y = []

def init():
    x.append(length/np.sqrt(2))
    y.append(-length/np.sqrt(2))
    
    pen_x.set_data(x[0],y[0])
    rod.set_data([0,x[0]],[0,y[0]])
    return pen_x, rod,

def animate(i):
    
    x_disp = x[i]
    y_disp = y[i]
    
    Î¸ = abs(np.arctan(x_disp/y_disp))
    accel = m * g * np.sin(Î¸)
    
    v_x = accel * np.cos(Î¸) * dt
    v_y = -accel * np.sin(Î¸) * dt 
    
    x[i] += v_x * dt
    y[i] -= v_y * dt
    
    x_disp = x[i]
    y_disp = y[i]
    
    Î¸ = abs(np.arctan(x_disp/y_disp))
    
    x.append(length * np.sin(Î¸))
    y.append(length * np.cos(Î¸))
    
    pen_x.set_data(x[i],y[i])
    rod.set_data([0,x[i]],[0,y[i]])

    return pen_x, rod, 

anim = FuncAnimation(fig, 
                     animate, 
                     frames=40,
                     init_func = init,
                     interval=100,
                     blit=True)

plt.show()
```
